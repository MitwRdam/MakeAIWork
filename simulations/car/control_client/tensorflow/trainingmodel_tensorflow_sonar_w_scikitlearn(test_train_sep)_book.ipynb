```python

import tensorflow as tf
keras = tf.keras
from keras import Model

import pandas as pd
import numpy as np

# Make numpy values easier to read.
np.set_printoptions(precision=4, suppress=True)

from sklearn.model_selection import train_test_split
```

```python
filename = "../sonar_kevin.csv"
```

```python
sonar_01_train = pd.read_csv(filename, sep = ',')
```

```python
sonar_01_labels = sonar_01_train.iloc[:, [3]]
sonar_01_sensors = sonar_01_train.iloc[:,[0, 1, 2]] # met .loc kan je labels gebruiken 'Sensor 1' etc.
```

```python
# sonar_01_labels.head()
```

```python
# sonar_01_sensors.head()
```

```python
sonar_01_sensors = np.array(sonar_01_sensors)
sonar_01_labels = np.array(sonar_01_labels)
```

```python
# print(sonar_01_sensors)
```

```python
# val_sonar_01_sensors = np.array(sonar_01_sensors [500:799])
# val_sonar_01_labels = np.array(sonar_01_labels [500:799])
```

```python
sonar_01_sensors_train, sonar_01_sensors_test, sonar_01_labels_train, sonar_01_labels_test = train_test_split(sonar_01_sensors, sonar_01_labels, test_size = 0.20)
```

```python
normalize = keras.layers.Normalization()
normalize.adapt(sonar_01_sensors_train) 
normalize.adapt(sonar_01_sensors_test)
normalize.adapt(sonar_01_labels_train) 
normalize.adapt(sonar_01_labels_test)
```

```python
print(sonar_01_sensors_train) 
print()
print(sonar_01_sensors_test)
print()
print(sonar_01_labels_train)
print() 
print(sonar_01_labels_test)
```

```python
norm_sonar_01_model = tf.keras.Sequential([
  normalize,
  keras.Input(shape = (3,)),
  keras.layers.Dense(16, activation = 'relu'), 
  keras.layers.Dense(16, activation = 'relu'),
  keras.layers.Dense(1)
])  
```

```python
norm_sonar_01_model.compile(loss = tf.keras.losses.MeanSquaredError(), metrics = ['accuracy', 'mae'], 
                            optimizer = tf.keras.optimizers.Adam(learning_rate=0.01))
```

```python
norm_sonar_01_model.summary()
```

```python
norm_sonar_01_model.fit(sonar_01_sensors_train, sonar_01_labels_train, epochs=100)
```

```python
norm_sonar_01_model.evaluate(sonar_01_sensors_test, sonar_01_labels_test)
```

```python
norm_sonar_01_model.predict(sonar_01_sensors_test)
```
